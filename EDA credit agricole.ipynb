### Data cleaning

```python
## on supprime les espaces pour éviter de consider deux meme valeurs comme différentes)
df['Last_Credit'] = df['Last_Credit'].str.strip()
```

```python
df['Market&Env_Context'] = df['Market&Env_Context'].str.strip()
```

```python
# * combine veuf with celibetaire (the two classes are married and not married)

df['Mrt_St'] = df['Mrt_St'].apply(lambda x: 'married' if x == 'MARIE' else 'not married')

# * for "Oth_act", combine all categories except for "aucune"  (the two classes are "Oui" and "non"
df['Oth_Act'] = df['Oth_Act'].apply(lambda x: 'NON' if x == 'aucune' or x == 'RAS' else 'OUI')

# * Combine Garantie réelle et personnelle with  Garantie personnelle et Financière (relable the column "Garantie finanicere", and the two categGories are "Oui" and "non"
df['Guarantee_Type'] = df['Guarantee_Type'].apply(lambda x: 'OUI' if x == 'Garantie financière' else 'NON')
df.rename(columns={'Guarantee_Type' : 'Finantial_Guarantee'}, inplace = True)

# * For Type_cult, combine to two classes:  "maraîchage" and "cereale"
df['Type_Cult']=df['Type_Cult'].apply(lambda x:'cereale' if x == 'mais' or x == 'maîs' else 'maraîchage')

# * For area, combine all that are greater than 1 ha  (notice that 1 Ha and 01 ha should be combined)

df['Area'] = df['Area'].apply(lambda x: '01ha' if x == '01ha' or x == '1ha' else 'over 01ha')

## Last_credit to two classes

df['Last_Credit'] = df['Last_Credit'].apply(lambda x: 'impayés' if x == 'Impayés' else 'Non impayés')
```

```python
#* Remove Commit_col  (there is too few data)
df = df.drop(columns = 'Commit_Col')
```

```python
# * Remove all categores that have 100 the same
colonne_sup = ['Own_Area','Proof_Area', 'Proximity', 'First_Assessment', 'Second_Assessment', 'Final_Decision']
df = df.drop(columns = colonne_sup ) 
```
